```python
import sys
from pathlib import Path

cwd = Path.cwd()
PROJECT_ROOT = cwd.parent if cwd.name == "notebooks" else cwd
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.config import DATASET_DIR, DATA_YAML

print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATASET_DIR:", DATASET_DIR)
print("DATA_YAML exists:", DATA_YAML.exists())
```

```python
def show_tree(root: Path, max_depth=4):
    root = root.resolve()
    for p in sorted(root.rglob("*")):
        rel = p.relative_to(root)
        if len(rel.parts) <= max_depth:
            print(rel, "/" if p.is_dir() else "")

show_tree(DATASET_DIR, max_depth=4)
```

```python
import shutil

valid_dir = DATASET_DIR / "valid"
val_dir = DATASET_DIR / "val"

if valid_dir.exists() and not val_dir.exists():
    shutil.move(str(valid_dir), str(val_dir))
    print("✅ Renamed valid/ -> val/")
else:
    print("No rename needed (either val/ exists already, or valid/ not present).")
```

```python
#verify image label counts match
def count_files(path: Path, exts=None):
    if not path.exists():
        return 0
    if exts is None:
        return len([p for p in path.iterdir() if p.is_file()])
    exts = tuple(e.lower() for e in exts)
    return len([p for p in path.iterdir() if p.is_file() and p.suffix.lower() in exts])

train_images = DATASET_DIR / "train/images"
train_labels = DATASET_DIR / "train/labels"
val_images = DATASET_DIR / "val/images"
val_labels = DATASET_DIR / "val/labels"

print("Train images:", count_files(train_images, exts=[".jpg", ".jpeg", ".png"]))
print("Train labels:", count_files(train_labels, exts=[".txt"]))
print("Val images:", count_files(val_images, exts=[".jpg", ".jpeg", ".png"]))
print("Val labels:", count_files(val_labels, exts=[".txt"]))
```

```python
#normalizing data.yaml paths(training error ghatauna)
import yaml

if not DATA_YAML.exists():
    raise FileNotFoundError(f"❌ data.yaml not found at {DATA_YAML}")

data = yaml.safe_load(DATA_YAML.read_text(encoding="utf-8"))

# Force YOLO to use this folder as the dataset root
data["path"] = str(DATASET_DIR)
data["train"] = "train/images"
data["val"] = "val/images"

# Optional: test split
if (DATASET_DIR / "test/images").exists():
    data["test"] = "test/images"

DATA_YAML.write_text(yaml.dump(data, sort_keys=False), encoding="utf-8")

print("✅ Updated data.yaml to safe relative paths:")
print(DATA_YAML.read_text(encoding="utf-8"))
```
